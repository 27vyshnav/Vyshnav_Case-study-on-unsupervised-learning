```python
#checking the missing value
from sklearn.preprocessing import StandardScaler
print(data.isnull().sum())
```

```python
X = data[['Flavanoids', 'Nonflavanoid_Phenols']]
X.head()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='Flavanoids',y='Nonflavanoid_Phenols',data=X,s=20)
plt.xlabel=('Flavanoids')
plt.ylabel=('Nonflavanoid_Phenols')
plt.title=('Flavanoids VS Nonflavanoid_Phenols')
plt.show()
```

```python
from sklearn.cluster import KMeans


wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter=300,
                 n_init = 10, random_state = 42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
plt.figure(figsize=(6,6))
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="o")
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel("Wcss")
plt.xticks(np.arange(1,11,1))
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5 , init = 'k-means++', max_iter=300 ,
                  n_init =10 , random_state =42)
kmeans.fit(X)

cluster_labels = kmeans.predict(X)

cluster_labels
```

```python
data1 = data.copy()

data1['Cluster'] =cluster_labels
data1.head()
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    score = silhouette_score(data_scaled, kmeans.labels_)
    silhouette_scores.append(score)

```

```python
silhouette_scores
```

```python
#DBSCAN
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data_scaled)

dbscan_labels
```
